```python
url = "https://aerodatabox.p.rapidapi.com/flights/airports/icao/EHAM/2022-08-03T05:00/2022-08-03T16:00"

querystring = {"withLeg":"true","withCancelled":"true","withCodeshared":"true","withCargo":"true","withPrivate":"true","withLocation":"false"}

headers = {
"X-RapidAPI-Key": "724145dec7msh1181498466f1f55p19b908jsn482bf314b0c4",
"X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
}

response = requests.get( url, headers=headers,params=querystring)
print(response.text)

# response = requests.request("GET", url, headers=headers)
```

```python
r_json=response.json()
r_json
```

```python
df_departures=pd.json_normalize(r_json['departures'])
pd.set_option('display.max_columns', None)
```

```python
df_departures.shape
```

```python
dept_cols=['number','departure.scheduledTimeUtc','departure.terminal','departure.gate','airline.name','arrival.airport.icao','arrival.airport.name','arrival.scheduledTimeUtc']
df_departures1=df_departures[dept_cols]
df_departures1.head(2)
```

```python
df_arrivals=pd.json_normalize(r_json['arrivals'])
df_arrivals.head(2)
```

```python
arriv_cols=['number','departure.airport.icao','departure.airport.name','arrival.scheduledTimeUtc','airline.name','departure.scheduledTimeUtc','departure.gate']
df_arrivals1=df_arrivals[arriv_cols]
df_arrivals1.shape
```

```python
df_dept_arr= pd.merge(df_arrivals1, df_departures1, how="inner", on=["number","airline.name"])
df_dept_arr.shape
```

```python
df_dept_arr['number']=df_dept_arr['number'].str.replace(" ","")
```

```python
df_dept_arr['number']
```

```python
df_dept_arr['number'].value_counts(ascending=False)
```

```python
cols_drop=["arrival.scheduledTimeUtc_x","arrival.scheduledTimeUtc_y","departure.gate_x","departure.gate_y","departure.scheduledTimeUtc_x","departure.scheduledTimeUtc_y"]
df_dept_arr.drop(columns=cols_drop,inplace=True)
```

**NOTE**:We have made an inner join in order to get the information of each flight from two different dataframes because the location of the location of the departures was missing in the "departures" dataframe (extraction). That is why a second extraction from the arrivals was made, so we could collect this information about the departures.
